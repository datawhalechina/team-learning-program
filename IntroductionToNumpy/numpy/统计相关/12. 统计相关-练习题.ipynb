```python
from numpy . linalg import inv 
from numpy import dot, transpose 
X = [[1, 6, 2] , [1, 8, 1] , [1, 10, 0] , [1 , 14, 2] , [1, 18, 0]] 
y = [[7] , [9] , [13] , [17.5] , [18]] 
```

```python
print(X)
print(y)
```

```python
print(dot(inv(dot(transpose(X) , X)) , dot(transpose(X) , y))) 
```

```python
import numpy as np

np.random.seed(100)
a = np.random.randint(1, 10, [5, 3])
print(a)
# [[9 9 4]
#  [8 8 1]
#  [5 3 6]
#  [3 3 3]
#  [2 1 9]]

b = np.amax(a, axis=1)
print(b)
# [9 8 6 3 9]
```
